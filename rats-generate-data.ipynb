```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

N = 100
np.random.seed(1)

x = np.random.uniform(low=1, high=7, size=(N,))
y = - ((x-5)**2) +25 + np.random.uniform(low=-3, high=3, size=(N,))
```

```python
plt.plot(x, y, "b.");
plt.xlabel("weight [dag]");
plt.ylabel("size [cm]");
plt.title("Rats data set");
```

```python
df = pd.DataFrame(data={"weight": x, "size": y})
#df.to_csv('datasets/rats-obesity-100examples.csv', sep = ',', index = False)

rats = pd.read_csv("datasets/rats-obesity-100examples.csv")
```
